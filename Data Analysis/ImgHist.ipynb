### Image Histogram

- Helps identify focused area of the image along with shadow information during the process of data collection

```python
import seaborn_image as isns
import seaborn as sns
from PIL import Image, ImageOps
import cv2

# set context 
isns.set_context("notebook")

# set global image settings
isns.set_image(cmap="deep", origin="lower")

# load sample dataset
img = cv2.imread('GESTURE DATASET/test/back_right/80.jpg')

print(img.shape)

	
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img.shape)

img = cv2.rotate(img, cv2.ROTATE_180)
fig = isns.imghist(img, dx=0.1, units="nm", cbar_label="Height (nm)", robust=True,  # set robust plotting
    perc=(20, 35))  # set the percentile of the data to view)
```

```python
ax = isns.imgplot(
    img,
    robust=True,  # set robust plotting
    perc=(20, 35),  # set the percentile of the data to view
    cbar_label="Height (nm)",
)
```
